```python
ct_path = r'C:\Users\roseh\Desktop\NYBG_R\data\archive\QC2\dnu'
data_path = r'C:\Users\roseh\Desktop\NYBG_R\data'
```

```python
main = pd.read_csv(os.path.join(data_path,'nybg_all_sentinelndvi.csv'),parse_dates=['date'])
```

```python
ct_628 = pd.read_csv(os.path.join(ct_path,'20220628_NYBG_7810.csv'),parse_dates=['Date'])
ct_628['Day'] = ct_628['Date'].dt.date.astype('datetime64[ns]')
```

```python
ct_628.replace('BG_UL3_v2','BG_ML3',inplace=True)
ct_628.replace('BE_SV3','BE_UL3',inplace=True)
ct_628.replace('HL','ML',inplace=True)
```

```python
ct_84['Day'] = ct_84['Date'].dt.date.astype('datetime64[ns]')
```
