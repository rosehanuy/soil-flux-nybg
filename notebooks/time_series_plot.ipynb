```python
import pandas as pd
import os
```

```python
import matplotlib.pyplot as plt
```

#### Create time series plots of NYBG and LDEO Data

```python
basepath = 'C:/Users/roseh/Desktop/NYBG_R/'
file_location = 'data/processed'
os.chdir(os.path.join(basepath, file_location))
```

```python
nybg = pd.read_csv('NYBG_means_all_dates.csv',usecols=['date','type','mean'],parse_dates=['date'])   # read in csv
nybg = nybg.pivot(index='date',columns='type',values='mean')                                         # create pivot table

ax = nybg.plot(figsize=(15,7),title='NYBG Mean CO2 Flux, 5/31/22 - 9/15/22')                              # plot table
ax.set_ylabel('Mean CO2')
plt.show()
```

```python
output_folder = 'C:/Users/roseh/Desktop/NYBG_R/reports/figures'
filename = 'nybg_time_series.png'
ax.figure.savefig(os.path.join(output_folder,filename))
```

```python
ldeo = pd.read_csv('LDEO_means_all_dates.csv',usecols=['date','type','mean'],parse_dates=['date'])   # read in csv
ldeo = ldeo.pivot(index='date',columns='type',values='mean')                                         # create pivot table

ax = ldeo.plot(figsize=(15,7),title='LDEO Mean CO2 Flux, 6/10/22 - 9/16/22')                              # plot table
ax.set_ylabel('Mean CO2')
plt.show()
```

```python
output_folder = 'C:/Users/roseh/Desktop/NYBG_R/reports/figures'
filename = 'ldeo_time_series.png'
ax.figure.savefig(os.path.join(output_folder,filename))
```

```python
fig,(ax1,ax2) = plt.subplots(2,sharex=True,figsize=(15,9))

nybg.plot(ax=ax1)

ldeo.plot(ax=ax2)  

plt.show()
```
