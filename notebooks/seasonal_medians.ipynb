```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

### Create Seasonal Median df - all temps, lst, ndvi and evi

```python
data_path = r'C:\Users\roseh\Desktop\NYBG_R\data'
```

```python
lst_df = pd.read_csv(os.path.join(data_path,'lst_nybg_may_nov.csv'))
```

```python
nybg_df = pd.read_csv(os.path.join(data_path,'nybg_all_main.csv'))
```

```python
ndvi_df = pd.read_csv(os.path.join(data_path,'ndvi_sentinel_nybg_may_nov.csv'))
```

```python
evi_df = pd.read_csv(os.path.join(data_path,'evi_may_nov_nybg.csv'))
```

```python
evi_median = evi_df.median(numeric_only=True)
evi_median_df = pd.DataFrame(evi_median)
evi_median_df = evi_median_df.reset_index()
evi_median_df = evi_median_df.rename(columns={'index':'labels',0:'evi'})
```

```python
evi_median_df
```

```python
nybg_df['date'] = pd.to_datetime(nybg_df['date'])
```

```python
nybg_df = nybg_df.loc[nybg_df['date'] < '2022-11-01']
```

```python
lst_df.columns
```

```python
lst_median = lst_df.median(numeric_only=True)
lst_median_df = pd.DataFrame(lst_median)
lst_median_df = lst_median_df.reset_index()
lst_median_df = lst_median_df.rename(columns={'index':'labels',0:'lst'})
lst_median_df
```

```python
ndvi_median = ndvi_df.median(numeric_only=True)
ndvi_median_df = pd.DataFrame(ndvi_median)
ndvi_median_df = ndvi_median_df.reset_index()
ndvi_median_df = ndvi_median_df.rename(columns={'index':'labels',0:'ndvi'})
```

```python
ndvi_median_df
```

```python
grouped = nybg_df.groupby('Label')
temps = []
tempnames = []
for name, group in grouped:
    if name in ['BE_UL1', 'BL_ML1', 'BR_HL3', 'BT_FE2', 'BT_FE3',
       'BT_FI1', 'BT_FI3', 'BW_HL1', 'BE_UL3', 'BG_ML3']:
        tempnames.append(name)
        temps.append(group['Soil temperature'].median())
        
temp_medians = pd.DataFrame(data={'tempnames': tempnames,'soil_temps':temps})     
```

```python
temp_medians
```

```python
chambertemp=[]
chambernames=[]
for name, group in grouped:
    if name in ['BE_UL1', 'BL_ML1', 'BR_HL3', 'BT_FE2', 'BT_FE3',
       'BT_FI1', 'BT_FI3', 'BW_HL1', 'BE_UL3', 'BG_ML3']:
        chambernames.append(name)
        chambertemp.append(group['Chamber_Temp'].median())
```

```python
chamber_medians = pd.DataFrame(data={'chambernames': chambernames,'chambertemp':chambertemp})
```

```python
chamber_medians
```

```python
res=[]
resnames=[]
for name, group in grouped:
    if name in ['BE_UL1', 'BL_ML1', 'BR_HL3', 'BT_FE2', 'BT_FE3',
       'BT_FI1', 'BT_FI3', 'BW_HL1', 'BE_UL3', 'BG_ML3']:
        resnames.append(name)
        res.append(group['Res'].median())
```

```python
res_medians = pd.DataFrame(data={'resnames': resnames,'res':res})
```

```python
df = chamber_medians.merge(temp_medians,left_on='chambernames',right_on='tempnames')
```

```python
df
```

```python
merge1 = df.merge(lst_median_df,left_on='tempnames',right_on='labels')
```

```python
merge1
```

```python
merge2 = merge1.merge(res_medians,left_on='tempnames',right_on='resnames')
```

```python
merge3 = merge2.merge(ndvi_median_df,left_on='tempnames',right_on='labels')
```

```python
merge4 = merge3.merge(evi_median_df,left_on='tempnames',right_on='labels')
```

```python
merge4.to_csv(os.path.join(data_path,'seasonal_medians_may_nov.csv'))
```
