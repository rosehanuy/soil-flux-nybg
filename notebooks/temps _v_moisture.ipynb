#### For each collar, plot Flux v. Soil Moisture

```python
basepath = 'C:/Users/roseh/Desktop/NYBG_R/'
file_location = 'data/processed/QC2'
os.chdir(os.path.join(basepath, file_location))
```

```python
df = pd.read_csv('NYBG_CO2_Flux_and_Temps.csv')
```

```python
df.drop(22,axis=0,inplace=True)
```

```python
df.info()
```

```python
grouped = df.groupby('Collar')
```

```python
def plot_temps(collar):
    """Make plot of soil flux v. soil moisture and soil moisture v. soil temperature"""
    
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,3))
    plt.subplots_adjust(hspace=0.5)
    ax1.scatter(collar[1]['Moisture'],collar[1]['CO2_Flux'])
    ax1.set_title('{}: CO2 v. Soil Moisture'.format(collar[0]))
    ax1.set_xlabel('Soil Moisture')
    ax1.set_ylabel('CO2 Flux')
    ax2.scatter(collar[1]['Moisture'],collar[1]['Temperature'])
    ax2.set_title('{}: Soil Temperature v. Soil Moisture'.format(collar[0]))
    ax2.set_xlabel('Soil Moisture')
    ax2.set_ylabel('Soil Temperature')
    # filepath = 'C:/Users/roseh/Desktop/NYBG_R/data/processed/QC2/collar_plots'
    # filename = '{} chamber_soil_temp_plots.png'.format(collar[0])
    # plt.savefig(os.path.join(filepath,filename))
    plt.show()
```

```python
collar_list = [group for group in grouped]
```

```python
for collar in collar_list:
    plot_temps(collar)
```

#### For each land type, plot Flux v. Soil Moisture

```python
grouped_2 = df.groupby('Type')
```

```python
type_list = [group for group in grouped_2]
```

```python
for collar in type_list:
    plot_temps(collar)
```
