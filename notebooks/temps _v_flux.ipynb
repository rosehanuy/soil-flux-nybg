```python
def plot_temps(collar):
    """Make side-by-side plots of CO2 flux v. Soil Temperature 
    and CO2 flux v. Chamber Temperature"""
    
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,3))
    
    ax1.scatter(collar[1]['Temperature'],collar[1]['CO2_Flux'])
    ax1.set_title('{}: CO2 v. Soil Temperature'.format(collar[0]))
    ax2.scatter(collar[1]['Chamber_Temp'],collar[1]['CO2_Flux'])
    ax2.set_title('{}: CO2 v. Chamber Temperature'.format(collar[0]))
    
    filepath = 'C:/Users/roseh/Desktop/NYBG_R/data/processed/QC2'
    filename = '{} chamber_soil_temp_plots.png'.format(collar[0])
    plt.savefig(os.path.join(filepath,filename))
    plt.show()
```

```python
grouped_2 = df.groupby('Type')
```
